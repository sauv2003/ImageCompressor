```python
from PIL import Image
import os

def compress_image(input_path, output_path, quality=85, max_size=(1920, 1080)):
    # Open the image
    img = Image.open(input_path)

    # Convert to RGB if needed (JPEG doesn’t support alpha)
    if img.mode in ("RGBA", "P"):
        img = img.convert("RGB")

    # Resize if larger than max_size while maintaining aspect ratio
    img.thumbnail(max_size, Image.Resampling.LANCZOS)

    # Save optimized image
    img.save(output_path, "JPEG", optimize=True, quality=quality)

    # Print size difference
    original = os.path.getsize(input_path) / 1024
    compressed = os.path.getsize(output_path) / 1024
    print(f"Original: {original:.2f} KB → Compressed: {compressed:.2f} KB ({(compressed/original)*100:.1f}% of original size)")
```

```python
compress_image("input.jpg", "compressed.jpg", quality=10)
```

```python
import io
def auto_compress(input_path, output_path, target_kb=300):
    img = Image.open(input_path)
    if img.mode in ("RGBA", "P"):
        img = img.convert("RGB")

    q = 95
    while q > 10:
        buffer = io.BytesIO()
        img.save(buffer, "JPEG", optimize=True, quality=q)
        size_kb = buffer.tell() / 1024
        if size_kb <= target_kb:
            break
        q -= 5
    with open(output_path, "wb") as f:
        f.write(buffer.getvalue())
    print(f"Final Quality: {q}, Size: {size_kb:.2f} KB")
```

```python
auto_compress("input.jpg", "compressed.jpg", target_kb = 50)
```
